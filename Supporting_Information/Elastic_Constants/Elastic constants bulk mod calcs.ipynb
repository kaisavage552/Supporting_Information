```python
import numpy as np

# Pure Mg elastic constants (GPa)
C11 = 64.66
C12 = 24.14
C13 = 21.19
C33 = 69.62
C44 = 17.14

print("="*60)
print("BULK MODULUS CALCULATION - Pure Mg (Hexagonal)")
print("="*60)

# Voigt average (upper bound)
B_voigt = (2*C11 + C33 + 2*C12 + 4*C13) / 9

# Reuss average (lower bound)
C_squared = (C11 + C12)*C33 - 2*C13**2
M = C11 + C12 + 2*C33 - 4*C13
B_reuss = C_squared / M

# Hill average (arithmetic mean - most commonly used)
B_hill = (B_voigt + B_reuss) / 2

print(f"\nElastic Constants:")
print(f"  C11 = {C11:.2f} GPa")
print(f"  C12 = {C12:.2f} GPa")
print(f"  C13 = {C13:.2f} GPa")
print(f"  C33 = {C33:.2f} GPa")
print(f"  C44 = {C44:.2f} GPa")

print(f"\nBulk Modulus Calculations:")
print(f"  B (Voigt)  = {B_voigt:.2f} GPa  (upper bound)")
print(f"  B (Reuss)  = {B_reuss:.2f} GPa  (lower bound)")
print(f"  B (Hill)   = {B_hill:.2f} GPa   (recommended)")

print(f"\nComparison:")
print(f"  Experimental: ~35-37 GPa")
print(f"  MEAM:    {B_hill:.2f} GPa")
print(f"  Deviation:    {((B_hill-36)/36*100):.1f}%")
print("="*60)
```
