```python
"""
Convert Mg2Zn11.cif to P1 (no symmetry) LAMMPS data file
"""
from pymatgen.core import Structure
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.io.lammps.data import LammpsData

# Read CIF
structure = Structure.from_file("Mg2Zn11.cif")

# Get conventional structure
sga = SpacegroupAnalyzer(structure, symprec=1e-3)
structure_full = sga.get_conventional_standard_structure()

# Convert to P1 (no symmetry)
structure_p1 = Structure(
    lattice=structure_full.lattice,
    species=structure_full.species,
    coords=structure_full.cart_coords,
    coords_are_cartesian=True
)

# Write LAMMPS data file
lammps_data = LammpsData.from_structure(structure_p1, atom_style='atomic')
lammps_data.write_file("mg2zn11_p1.data")

print(f"Done. Output: mg2zn11_p1.data ({len(structure_p1)} atoms)")
```
